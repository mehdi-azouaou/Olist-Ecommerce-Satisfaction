## 5. Nettoyage du dataset customers

````python
# On crée une copie pour conserver le dataset brut intact.
customers_clean = customers.copy()

# customer_id est un identifiant : il doit rester en texte.
customers_clean["customer_id"] = customers_clean["customer_id"].astype(str)

# customer_unique_id est aussi un identifiant client : il doit rester en texte.
customers_clean["customer_unique_id"] = customers_clean["customer_unique_id"].astype(str)

# Le code postal est une information géographique, pas une quantité à calculer.
# On le transforme donc en texte et on conserve 5 caractères.
customers_clean["customer_zip_code_prefix"] = customers_clean["customer_zip_code_prefix"].astype(str).str.zfill(5)

# On uniformise les noms de villes.
customers_clean["customer_city"] = customers_clean["customer_city"].astype(str).str.lower().str.strip()

# Les États brésiliens sont codés sur deux lettres : on les met en majuscules.
customers_clean["customer_state"] = customers_clean["customer_state"].astype(str).str.upper().str.strip()

# On supprime uniquement les doublons complets.
customers_clean = customers_clean.drop_duplicates()

# Vérification.
customers_clean.info()
````

## 6. Nettoyage du dataset sellers

````python
# On crée une copie du dataset vendeurs.
sellers_clean = sellers.copy()

# seller_id est un identifiant : on le garde en texte.
sellers_clean["seller_id"] = sellers_clean["seller_id"].astype(str)

# Le code postal est une variable géographique : on le garde en texte avec 5 caractères.
sellers_clean["seller_zip_code_prefix"] = sellers_clean["seller_zip_code_prefix"].astype(str).str.zfill(5)

# On uniformise les villes.
sellers_clean["seller_city"] = sellers_clean["seller_city"].astype(str).str.lower().str.strip()

# On uniformise les États.
sellers_clean["seller_state"] = sellers_clean["seller_state"].astype(str).str.upper().str.strip()

# Suppression des doublons complets éventuels.
sellers_clean = sellers_clean.drop_duplicates()

# Vérification.
sellers_clean.info()
````

## 7. Nettoyage du dataset orders

````python
# On crée une copie du dataset commandes.
orders_clean = orders.copy()

# order_id est l'identifiant de la commande : on le garde en texte.
orders_clean["order_id"] = orders_clean["order_id"].astype(str)

# customer_id est la clé permettant de relier orders à customers.
orders_clean["customer_id"] = orders_clean["customer_id"].astype(str)

# order_status est une variable catégorielle : on uniformise le texte.
orders_clean["order_status"] = orders_clean["order_status"].astype(str).str.lower().str.strip()

# Liste des colonnes qui contiennent des dates.
order_date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

# Conversion des colonnes de dates.
# errors="coerce" transforme les valeurs non convertibles en NaT.
for col in order_date_columns:
    orders_clean[col] = pd.to_datetime(orders_clean[col], errors="coerce")

# Les dates manquantes sont conservées.
# Exemple : une commande annulée peut ne pas avoir de date de livraison.
orders_clean = orders_clean.drop_duplicates()

# Vérification.
orders_clean.info()
````

## 8. Nettoyage du dataset order_items

````python
# On crée une copie du dataset des articles commandés.
order_items_clean = order_items.copy()

# order_id est une clé de jointure : on le garde en texte.
order_items_clean["order_id"] = order_items_clean["order_id"].astype(str)

# order_item_id représente le numéro de l'article dans la commande : on le garde en entier.
order_items_clean["order_item_id"] = order_items_clean["order_item_id"].astype(int)

# product_id est un identifiant : on le garde en texte.
order_items_clean["product_id"] = order_items_clean["product_id"].astype(str)

# seller_id est un identifiant : on le garde en texte.
order_items_clean["seller_id"] = order_items_clean["seller_id"].astype(str)

# shipping_limit_date est une date : on la convertit.
order_items_clean["shipping_limit_date"] = pd.to_datetime(order_items_clean["shipping_limit_date"], errors="coerce")

# price est un montant : on le garde en nombre décimal.
order_items_clean["price"] = order_items_clean["price"].astype(float)

# freight_value représente les frais de livraison : on le garde en nombre décimal.
order_items_clean["freight_value"] = order_items_clean["freight_value"].astype(float)

# Suppression des doublons complets éventuels.
order_items_clean = order_items_clean.drop_duplicates()

# Vérification.
order_items_clean.info()
````

## 9. Nettoyage du dataset order_payments

````python
# On crée une copie du dataset des paiements.
order_payments_clean = order_payments.copy()

# order_id est une clé de jointure : on le garde en texte.
order_payments_clean["order_id"] = order_payments_clean["order_id"].astype(str)

# payment_sequential est un ordre de paiement : on le garde en entier.
order_payments_clean["payment_sequential"] = order_payments_clean["payment_sequential"].astype(int)

# payment_type est une catégorie : on uniformise le texte.
order_payments_clean["payment_type"] = order_payments_clean["payment_type"].astype(str).str.lower().str.strip()

# not_defined signifie que le type de paiement n'est pas connu.
# On le remplace par unknown, plus explicite.
order_payments_clean["payment_type"] = order_payments_clean["payment_type"].replace("not_defined", "unknown")

# payment_installments représente le nombre de mensualités : on le garde en entier.
order_payments_clean["payment_installments"] = order_payments_clean["payment_installments"].astype(int)

# Une mensualité égale à 0 est peu logique pour un paiement.
# On la remplace par 1 pour indiquer un paiement en une fois.
order_payments_clean["payment_installments"] = order_payments_clean["payment_installments"].replace(0, 1)

# payment_value est un montant : on le garde en nombre décimal.
order_payments_clean["payment_value"] = order_payments_clean["payment_value"].astype(float)

# Suppression des doublons complets éventuels.
order_payments_clean = order_payments_clean.drop_duplicates()

# Vérification.
order_payments_clean.info()
````

## 10. Nettoyage du dataset order_reviews

````python
# On crée une copie du dataset des avis clients.
order_reviews_clean = order_reviews.copy()

# review_id est un identifiant : on le garde en texte.
order_reviews_clean["review_id"] = order_reviews_clean["review_id"].astype(str)

# order_id est une clé de jointure : on le garde en texte.
order_reviews_clean["order_id"] = order_reviews_clean["order_id"].astype(str)

# review_score est une note de 1 à 5 : on le garde en entier.
order_reviews_clean["review_score"] = order_reviews_clean["review_score"].astype(int)

# Beaucoup de clients laissent une note sans titre de commentaire.
# On ne supprime pas ces lignes car la note reste utile pour analyser la satisfaction.
order_reviews_clean["review_comment_title"] = order_reviews_clean["review_comment_title"].fillna("no_comment")

# Même logique pour le message du commentaire.
order_reviews_clean["review_comment_message"] = order_reviews_clean["review_comment_message"].fillna("no_comment")

# Conversion de la date de création de l'avis.
order_reviews_clean["review_creation_date"] = pd.to_datetime(order_reviews_clean["review_creation_date"], errors="coerce")

# Conversion de la date de réponse à l'avis.
order_reviews_clean["review_answer_timestamp"] = pd.to_datetime(order_reviews_clean["review_answer_timestamp"], errors="coerce")

# Suppression des doublons complets éventuels.
order_reviews_clean = order_reviews_clean.drop_duplicates()

# Vérification.
order_reviews_clean.info()
````

## 11. Nettoyage du dataset products

````python
# On crée une copie du dataset produits.
products_clean = products.copy()

# Les colonnes product_name_lenght et product_description_lenght contiennent une faute d'orthographe.
# On les renomme avec length pour avoir des noms plus propres.
products_clean = products_clean.rename(columns={
    "product_name_lenght": "product_name_length",
    "product_description_lenght": "product_description_length"
})

# product_id est un identifiant : on le garde en texte.
products_clean["product_id"] = products_clean["product_id"].astype(str)

# Les catégories manquantes sont remplacées par unknown.
# On évite de supprimer les produits car ils peuvent apparaître dans des commandes.
products_clean["product_category_name"] = products_clean["product_category_name"].fillna("unknown")

# On uniformise les catégories.
products_clean["product_category_name"] = products_clean["product_category_name"].astype(str).str.lower().str.strip()

# Si la longueur du nom produit est manquante, on remplace par 0.
products_clean["product_name_length"] = products_clean["product_name_length"].fillna(0).astype(int)

# Même logique pour la longueur de description.
products_clean["product_description_length"] = products_clean["product_description_length"].fillna(0).astype(int)

# Si le nombre de photos est manquant, on remplace par 0.
products_clean["product_photos_qty"] = products_clean["product_photos_qty"].fillna(0).astype(int)

# Colonnes numériques décrivant les dimensions physiques des produits.
product_dimension_columns = [
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

# Pour les dimensions physiques manquantes, on remplace par la médiane.
# La médiane est moins sensible aux valeurs extrêmes que la moyenne.
for col in product_dimension_columns:
    median_value = products_clean[col].median()
    products_clean[col] = products_clean[col].fillna(median_value).astype(float)

# Suppression des doublons complets éventuels.
products_clean = products_clean.drop_duplicates()

# Vérification.
products_clean.info()
````

## 12. Nettoyage du dataset translation

````python
# On crée une copie du dataset de traduction.
translation_clean = translation.copy()

# On uniformise la catégorie en portugais.
translation_clean["product_category_name"] = translation_clean["product_category_name"].astype(str).str.lower().str.strip()

# On uniformise la catégorie en anglais.
translation_clean["product_category_name_english"] = translation_clean["product_category_name_english"].astype(str).str.lower().str.strip()

# Suppression des doublons complets éventuels.
translation_clean = translation_clean.drop_duplicates()

# Vérification.
translation_clean.info()
````

## 13. Nettoyage du dataset geolocation

````python
# On crée une copie du dataset de géolocalisation.
geolocation_clean = geolocation.copy()

# Le code postal est une variable géographique : on le garde en texte avec 5 caractères.
geolocation_clean["geolocation_zip_code_prefix"] = geolocation_clean["geolocation_zip_code_prefix"].astype(str).str.zfill(5)

# Latitude et longitude sont des coordonnées numériques.
geolocation_clean["geolocation_lat"] = geolocation_clean["geolocation_lat"].astype(float)
geolocation_clean["geolocation_lng"] = geolocation_clean["geolocation_lng"].astype(float)

# On uniformise les noms de villes.
geolocation_clean["geolocation_city"] = geolocation_clean["geolocation_city"].astype(str).str.lower().str.strip()

# On uniformise les États.
geolocation_clean["geolocation_state"] = geolocation_clean["geolocation_state"].astype(str).str.upper().str.strip()

# On supprime les doublons complets.
geolocation_clean = geolocation_clean.drop_duplicates()

# Le fichier geolocation contient souvent plusieurs lignes pour le même code postal.
# On agrège donc les coordonnées par code postal, ville et État.
geolocation_clean = (
    geolocation_clean
    .groupby(
        ["geolocation_zip_code_prefix", "geolocation_city", "geolocation_state"],
        as_index=False
    )
    .agg({
        "geolocation_lat": "mean",
        "geolocation_lng": "mean"
    })
)

# Vérification.
geolocation_clean.info()
````

## 14. Vérification finale

````python
# On rassemble les datasets nettoyés.
clean_datasets = {
    "customers_clean": customers_clean,
    "geolocation_clean": geolocation_clean,
    "order_items_clean": order_items_clean,
    "order_payments_clean": order_payments_clean,
    "order_reviews_clean": order_reviews_clean,
    "orders_clean": orders_clean,
    "products_clean": products_clean,
    "sellers_clean": sellers_clean,
    "translation_clean": translation_clean
}

# On vérifie les dimensions, les valeurs manquantes et les doublons après nettoyage.
for name, df in clean_datasets.items():
    print("=" * 80)
    print("DATASET NETTOYÉ :", name)
    print("=" * 80)

    print("\nDimensions :")
    print(df.shape)

    print("\nValeurs manquantes restantes :")
    print(df.isna().sum())

    print("\nDoublons complets restants :")
    print(df.duplicated().sum())

    print("\n")
````
